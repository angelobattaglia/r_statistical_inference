```r
# I generate a population from a Normal distribution
set.seed(42)  # for reproducibility
population <- rnorm(10000, mean = 50, sd = 10)  # a large population
```

```r
# Taking a sample from the population
sample <- sample(population, size = 100)  # campione di 100 individui
```

```r
# Evaluating the mean and the variance of the sample
sample_mean <- mean(sample)
sample_variance <- var(sample)

sample_mean
sample_variance
```

```r
# Sampling distribution of the mean
sample_means <- replicate(1000, mean(sample(population, size = 100)))
```

```r
# Histogram of sample means
hist(sample_means, breaks = 30, main = "Sampling Distribution of the Mean", xlab = "Sample Mean")

# Calculate a 95% confidence interval for the population mean
conf_interval <- t.test(sample)$conf.int

conf_interval

# One-sample t-test
t.test(sample, mu = 50)

# if not installed, run this on you R-console: install.packages("ggplot2")
library(ggplot2)

# Create a data frame for plotting
df <- data.frame(sample_means)

# Plot the sampling distribution with ggplot2
p <- ggplot(df, aes(x = sample_means)) +
  geom_histogram(aes(y = ..density..), binwidth = 0.2, colour = "black", fill = "white") +
  geom_density(alpha = .2, fill = "#FF6666") +
  labs(title = "Sampling Distribution of the Mean", x = "Sample Mean", y = "Density")

print(p)

## Saving it as an image (To fix)
#gsave("C:/Users/angel/Desktop/myplot.png", plot = p, width = 10, height = 8)
```
